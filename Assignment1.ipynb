## Question 3: Chat System using OOP

```python
from datetime import datetime

class User:
    """Class representing a user in the chat system"""
    def __init__(self, user_id, username):
        self.user_id = user_id
        self.username = username
    
    def __repr__(self):
        return f"User({self.username})"


class Message:
    """Class representing a message in the chat system"""
    def __init__(self, sender, content):
        self.sender = sender
        self.content = content
        self.timestamp = datetime.now()
    
    def __repr__(self):
        return f"[{self.timestamp.strftime('%H:%M:%S')}] {self.sender.username}: {self.content}"


class ChatRoom:
    """Class representing a chat room"""
    def __init__(self, room_name):
        self.room_name = room_name
        self.users = {}  # Dictionary to store users: {user_id: User}
        self.chat_history = []  # List to store messages
    
    def add_user(self, user):
        """Add a user to the chat room"""
        if user.user_id not in self.users:
            self.users[user.user_id] = user
            print(f"✓ {user.username} joined the chat room '{self.room_name}'")
        else:
            print(f"✗ {user.username} is already in the chat room")
    
    def remove_user(self, user_id):
        """Remove a user from the chat room"""
        if user_id in self.users:
            user = self.users.pop(user_id)
            print(f"✓ {user.username} left the chat room '{self.room_name}'")
        else:
            print(f"✗ User not found in the chat room")
    
    def send_message(self, user_id, content):
        """Send a message to the chat room"""
        if user_id not in self.users:
            print(f"✗ User is not in this chat room")
            return
        
        user = self.users[user_id]
        message = Message(user, content)
        self.chat_history.append(message)
        print(f"✓ Message sent: {message}")
    
    def view_chat_history(self):
        """Display the chat history"""
        print(f"\n--- Chat History for '{self.room_name}' ---")
        if not self.chat_history:
            print("No messages yet.")
        else:
            for message in self.chat_history:
                print(message)
        print("---" + "-" * (len(self.room_name) + 22) + "---\n")
    
    def get_active_users(self):
        """Get list of active users in the chat room"""
        return list(self.users.values())


# Example Usage
print("=== Chat System Demo ===\n")

# Create users
user1 = User(1, "Ali")
user2 = User(2, "Bilal")
user3 = User(3, "Tipu")

# Create a chat room
room = ChatRoom("General")

# Users join the chat room
room.add_user(user1)
room.add_user(user2)
room.add_user(user3)

print(f"\nActive users: {[u.username for u in room.get_active_users()]}\n")

# Send messages
room.send_message(1, "Hello everyone!")
room.send_message(2, "Hi Ali, how are you?")
room.send_message(3, "Hey guys!")
room.send_message(1, "I'm doing great!")

# View chat history
room.view_chat_history()

# User leaves the chat room
room.remove_user(2)
print(f"Active users: {[u.username for u in room.get_active_users()]}\n")

# Continue sending messages
room.send_message(3, "Bilal left, see you later!")
room.view_chat_history()
```
